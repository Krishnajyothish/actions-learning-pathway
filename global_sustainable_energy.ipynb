```python
# Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 1. load the data set
gdse_df = pd.read_csv("global-data-sustainable-energy.csv")
```

```python
# 2. Display the first 20 rows.
first_20_rows = gdse_df[:20]
print(first_20_rows)
```

```python
# 3.Find the number of null values in the dataset.
missing_values = gdse_df.isnull().sum()
print("Missing values:\n", missing_values)
```

```python
from sklearn.impute import SimpleImputer
```

```python
# 4. Impute the missing values with the mean values of the data. You can use SimpleImputer from sklearn.impute.
columns_to_impute = ['Access to electricity (% of population)','Access to clean fuels for cooking','Renewable-electricity-generating-capacity-per-capita','Financial flows to developing countries (US $)','Renewable energy share in the total final energy consumption (%)','Electricity from fossil fuels (TWh)','Electricity from nuclear (TWh)',
                                      'Electricity from renewables (TWh)','Low-carbon electricity (% electricity)','Primary energy consumption per capita (kWh/person)','Energy intensity level of primary energy (MJ/$2017 PPP GDP)','Value_co2_emissions_kt_by_country','Renewables (% equivalent primary energy)','gdp_growth','gdp_per_capita'
                                      ,'Land Area(Km2)','Latitude','Longitude']
```

```python
mean_values= gdse_df.groupby('Entity')[columns_to_impute].mean()
mean_value_filled=gdse_df[columns_to_impute].fillna(mean_values)
```

```python
print(mean_value_filled)
```

```python
# 5. Find the count, mean, standard deviation, quartiles, and extrema for the numeric columns.
mean_value_filled.describe()
```

```python
# 6. Create a new column, northern_hem. For countries in the northern hemisphere,northern_hem = 1. Else, northern_hem = 0.
# Hint: you can (should) generate these values by looking at the Latitude column.
mean_value_filled ['northern_hem'] = (mean_value_filled ['Latitude'] > 0).astype(int)
```

```python
print(mean_value_filled.head(5))
```

```python
# 7. Find the correlation between the numeric columns and display your findings on a heatmap.
number_cols = mean_value_filled.select_dtypes(include=['number'])
```

```python
# 8. Plot a histogram of the Low-carbon electricity (% electricity).
sns.histplot(mean_value_filled['Low-carbon electricity (% electricity)'], kde=True, bins=15)
plt.title('Histogram of Low-carbon Electricity (% electricity)')
plt.xlabel('Low-carbon Electricity (% electricity)')
plt.ylabel('Frequency')
plt.show()
```

```python
# 9. Compare the Value_co2_emissions_kt_by_country in countries in the northern hemisphere to that in the southern hemisphere using violin plots next to each other.
sns.violinplot(x='northern_hem', y='Value_co2_emissions_kt_by_country', data=mean_value_filled)
plt.title('CO2 Emissions by Hemisphere')
plt.xlabel('Southern & Northern Hemisphere')
plt.ylabel('CO2 Emissions (kt)')
plt.show()
```

```python
merged_df = pd.concat([mean_value_filled, gdse_df[['Entity','Year']]], axis=1)
```

```python
# 10. On the same line plot, display the Renewable-electricity-generating-capacity-per-capita from 2000 to 2015 for Argentina, Brazil and Chile.
df_filtered = merged_df[((merged_df['Entity'] == 'Argentina') | (merged_df['Entity'] == 'Brazil') | (merged_df['Entity']== 'Chile')) & (merged_df['Year'].between(2000, 2015))]
```

```python
#Create a line plot
sns.lineplot(x='Year', y='Renewable-electricity-generating-capacity-per-capita', hue='Entity', data=df_filtered)
plt.title('Renewable Energy Capacity per Capita (2000-2015)')
plt.xlabel('Year')
plt.ylabel('Renewable Energy Capacity per Capita')
plt.legend()
plt.show()
```

```python
# 11. Compare the average Electricity from fossil fuels (TWh) over the years against the average Electricity from renewables (TWh)
#over the years using a scatter plot for the following countries: Belgium, Brazil, Cameroon, Canada, China, France, Ghana, India,
# the United Kingdom and the United States of America. What can you conclude?
countries = ['Belgium', 'Brazil', 'Cameroon', 'Canada', 'China', 'France', 'Ghana', 'India', 'United Kingdom', 'United States']
compare_data = merged_df[merged_df['Entity'].isin(countries)][['Entity', 'Year', 'Electricity from fossil fuels (TWh)', 'Electricity from renewables (TWh)']]
avg_data = compare_data.groupby('Entity')[['Electricity from fossil fuels (TWh)', 'Electricity from renewables (TWh)']].mean()
plt.scatter(avg_data['Electricity from fossil fuels (TWh)'], avg_data['Electricity from renewables (TWh)'])
plt.xlabel('Electricity from Fossil Fuels (TWh)')
plt.ylabel('Electricity from Renewables (TWh)')
plt.title('Comparison of Average Fossil Fuels and Renewables Electricity')
plt.show()

# I conclude with the above scatterplot it has a postive relationship between
# Electricity from fossil fuels (TWh) over the years against the average Electricity from renewables (TWh) over the years.
```

```python
# 12. Plot the global trend of the total Electricity from fossil fuels (TWh) from 2000 to 2020
global_trend = merged_df.groupby('Year')['Electricity from fossil fuels (TWh)'].sum().reset_index()
plt.figure(figsize=(10, 6))
sns.lineplot(x='Year', y='Electricity from fossil fuels (TWh)', data=global_trend)
plt.title('Global Trend of Electricity from Fossil Fuels (2000-2020)')
plt.xlabel('Year')
plt.ylabel('Total Electricity from Fossil Fuels (TWh)')
plt.grid(True)
plt.show()
```

```python
# 13.On the same plot as part 12, plot the global trend of the total Electricity from fossil fuels (TWh) from 2000 to 2020 in:
# a. Countries in the northern hemisphere (1)
# b. Countries in the southern hemisphere (1)
# Do not forget to add labels to the plots.
merged_df ['southern_hem'] = (mean_value_filled ['Latitude'] < 0).astype(int)
northern_hem_data = merged_df[merged_df['northern_hem'] == True]
southern_hem_data = merged_df[merged_df['northern_hem'] == False]
northern_trend = northern_hem_data.groupby('Year')['Electricity from fossil fuels (TWh)'].sum().reset_index()
southern_trend = southern_hem_data.groupby('Year')['Electricity from fossil fuels (TWh)'].sum().reset_index()
plt.figure(figsize=(10,6))
sns.lineplot(x='Year', y='Electricity from fossil fuels (TWh)', data=northern_trend, label='Northern Hemisphere')
sns.lineplot(x='Year', y='Electricity from fossil fuels (TWh)', data=southern_trend, label='Southern Hemisphere')
plt.title('Global Trend of Electricity from Fossil Fuels by Hemisphere (2000-2020)')
plt.xlabel('Year')
plt.ylabel('Total Electricity from Fossil Fuels (TWh)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# In the year 2020
# a. Compare the Primary energy consumption per capita (kWh/person) with gdp_per_capita using a scatter plot.
# b. What is the value of the Pearson correlation coefficient?
# c. Draw the best regression line on the same plot as (a).
# d. What can you conclude?

from scipy.stats import pearsonr
data_2020 = merged_df[merged_df['Year'] == 2020]
data_2020 = data_2020.dropna(subset=['gdp_per_capita', 'Primary energy consumption per capita (kWh/person)'])
data_2020 = data_2020[np.isfinite(data_2020['gdp_per_capita'])]
data_2020 = data_2020[np.isfinite(data_2020['Primary energy consumption per capita (kWh/person)'])]
plt.figure(figsize=(10, 6))
sns.scatterplot(x='gdp_per_capita', y='Primary energy consumption per capita (kWh/person)', data=data_2020)
plt.title('Primary Energy Consumption vs. GDP per Capita (2020)')
plt.xlabel('GDP per Capita')
plt.ylabel('Primary Energy Consumption per Capita (kWh/person)')
correlation_coeff,p_value = pearsonr(data_2020['gdp_per_capita'], data_2020['Primary energy consumption per capita (kWh/person)'])
print(f"Pearson Correlation Coefficien is {correlation_coeff:.2f},  indicating a strong positive linear relationship.")
slope, intercept = np.polyfit(data_2020['gdp_per_capita'], data_2020['Primary energy consumption per capita (kWh/person)'], 1)
plt.plot(data_2020['gdp_per_capita'], slope * data_2020['gdp_per_capita'] + intercept, color='red', label='Regression Line')
plt.legend()
plt.show()
print("\nConclusion:")
print("There is a strong positive correlation between primary energy consumption per capita and GDP per capita in 2020.")
print(f"The Pearson correlation coefficient is {correlation_coeff:.2f}, indicating a strong positive linear relationship.")
print("This suggests that as GDP per capita increases, so does primary energy consumption per capita.")
print("The regression line visualizes this relationship, showing the general trend of increasing energy consumption with increasing GDP.")
```
